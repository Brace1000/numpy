```python
# Exercise 3: slicing
import numpy as np

# 1. Create a NumPy array of dimension 1 that contains all integers from 1 to 100 ordered.
numbers_1_to_100 = np.arange(1, 101)
print("1. Array from 1 to 100:")
print(numbers_1_to_100)

# 2. Without using a for loop, create an array that contains all odd integers.
odd_integers = numbers_1_to_100[::2] # Start from 1, step by 2
print("\n2. Odd integers:")
print(odd_integers)

# 3. Without using a for loop, create an array that contains all even integers reversed.
even_integers_reversed = numbers_1_to_100[1::2][::-1] # Get evens (start from index 1), then reverse
print("\n3. Even integers reversed:")
print(even_integers_reversed)

# 4. Using array of Q1, set the value of every 3 elements of the list (starting with the second) to 0.
modified_array = numbers_1_to_100.copy() # Make a copy to avoid modifying the original
modified_array[1::3] = 0 # Starting with the second element (index 1), every 3rd element
print("\n4. Array with every 3rd element (starting second) set to 0:")
print(modified_array)
```

```python
# Exercise 6: Broadcasting and Slicing
import numpy as np

# 1. Generate a 2-dimensional array of size 9x9, with all elements initialized to 1 and of type int8.
array_9x9_ones = np.ones((9, 9), dtype=np.int8)
print("1. 9x9 array of ones:")
print(array_9x9_ones)

# 2. Using slicing, create the specified array (a 'maze' like pattern)
# Start with a copy of the ones array
maze_array = np.ones((9, 9), dtype=np.int8)

# Set the inner square to zeros (from row 1 to 7, column 1 to 7)
maze_array[1:8, 1:8] = 0

# Create the inner 'paths' of ones
maze_array[2, 2:7] = 1
maze_array[3:6, 2] = 1
maze_array[3:6, 6] = 1
maze_array[4, 4] = 1
maze_array[6, 2:7] = 1

print("\n2. Maze-like array:")
print(maze_array)

# 3. Using broadcasting create an output matrix based on the following two arrays:
array_1 = np.array([1,2,3,4,5], dtype=np.int8)
array_2 = np.array([1,2,3], dtype=np.int8)

# Perform outer product multiplication for broadcasting
broadcasted_matrix = array_1[:, np.newaxis] * array_2 # Or just array_1.reshape(-1, 1) * array_2

print("\n3. Broadcasted matrix:")
print(broadcasted_matrix)
```

```python
# 4.Exercise 7: NaN

import numpy as np

generator = np.random.default_rng(123)
grades = np.round(generator.uniform(low = 0.0, high = 10.0, size = (10, 2)))
grades[[1,2,5,7], [0,0,0,0]] = np.nan
print("Original grades array:")
print(grades)

# Using np.where, create a third column that takes the grade of the first exam if available;
# otherwise, it uses the grade from the second exam. Add this column as the third column of the array.
third_column = np.where(np.isnan(grades[:, 0]), grades[:, 1], grades[:, 0])

# Reshape the third_column to be a 2D array with a single column to concatenate
third_column = third_column.reshape(-1, 1)

# Concatenate the new column to the grades array
grades_with_new_column = np.concatenate((grades, third_column), axis=1)

print("\nGrades array with new column:")
print(grades_with_new_column)
```

```python
# Exercise 8: Wine
import numpy as np
# 1. Load the data using genfromtxt, specifying the delimiter as ';' with excluding the headers, and optimize the numpy array size by reducing the data types. Use np.float32 and verify that the resulting numpy array weighs 76800 bytes.
data = np.genfromtxt("winequality-red.csv", delimiter=";", skip_header=1, dtype=np.float32)

print("Array shape:", data.shape)     
print("Array dtype:", data.dtype)      
print("Array size in bytes:", data.nbytes, "\n")  

# 2. Display the 2nd, 7th, and 12th rows as a two-dimensional array. Exclude np.nan values if present.
rows = data[[1, 6, 11], :]
rows_no_nan = rows[~np.isnan(rows).any(axis=1)]
np.set_printoptions(suppress=True, precision=3)
print(rows_no_nan, "\n")

# 3. Determine if there is any wine in the dataset with an alcohol percentage greater than 20%. Return True or False.
has_high_alcohol = np.any(data[:, -1] > 20)

print(has_high_alcohol, "\n")

#4. Calculate the average alcohol percentage across all wines in the dataset. Exclude np.nan values if present.
alcohol = data[:, 10]
avg_alcohol = np.nanmean(alcohol)
print(avg_alcohol, "\n")

#5. Compute various statistical measures (minimum, maximum, 25th percentile, 50th percentile, 75th percentile and the mean for the pH values).
ph_values = data[:, 8]
ph_values = ph_values[~np.isnan(ph_values)]
ph_min = np.min(ph_values)
ph_max = np.max(ph_values)
ph_25 = np.percentile(ph_values, 25)
ph_50 = np.percentile(ph_values, 50)   # same as median
ph_75 = np.percentile(ph_values, 75)
ph_mean = np.mean(ph_values)

print("Minimum:", ph_min)
print("Maximum:", ph_max)
print("25th Percentile:", ph_25)
print("50th Percentile:", ph_50)
print("75th Percentile:", ph_75)
print("Mean:", ph_mean, "\n")

#6.Find the average quality score of wines with the 20% least sulphate content.

sulphates = data[:, 9]
quality = data[:, 11]

mask = ~np.isnan(sulphates) & ~np.isnan(quality)
sulphates = sulphates[mask]
quality = quality[mask]

perc20 = np.percentile(sulphates, 20)

low_sulphates_mask = sulphates <= perc20

low_sulphates_quality = quality[low_sulphates_mask]

avg_quality_low_sulphates = np.mean(low_sulphates_quality)

print("Average quality (lowest 20% sulphates):", avg_quality_low_sulphates, "\n")

# 7.Compute the mean of all variables for wines with the best quality. Also, do the same for wines with the worst quality.

quality = data[:, 11]

mask = ~np.isnan(quality)
data_no_nan = data[mask]
quality_no_nan = quality[mask]

max_quality = np.max(quality_no_nan)
min_quality = np.min(quality_no_nan)

best_mask = quality_no_nan == max_quality
worst_mask = quality_no_nan == min_quality

mean_best_quality = np.mean(data_no_nan[best_mask], axis=0)
mean_worst_quality = np.mean(data_no_nan[worst_mask], axis=0)

print("Mean values for best quality wines:\n", mean_best_quality, "\n")

print("Mean values for worst quality wines:\n", mean_worst_quality)


```

```python
# Exercise 9: Football tournament

import numpy as np
import itertools


forecasts = np.genfromtxt("model_forecasts.txt", delimiter=",")

# Number of teams
n = forecasts.shape[0]

# All possible pairings (unordered, so (i,j) same as (j,i))
pairs = [(i, j) for i in range(n) for j in range(i+1, n)]

# Function to compute squared sum for a given pairing set
def score_for_matching(matching):
    diffs = [forecasts[i, j] for (i, j) in matching]
    return np.sum(np.square(diffs))

# Generate all possible perfect matchings using itertools
def generate_matchings(teams):
    if len(teams) == 0:
        return [[]]
    first = teams[0]
    matchings = []
    for i in range(1, len(teams)):
        for rest in generate_matchings(teams[1:i] + teams[i+1:]):
            matchings.append([(first, teams[i])] + rest)
    return matchings

# Generate all matchings for 10 teams
all_matchings = generate_matchings(list(range(n)))

# Find the best matching (minimum squared sum)
best_matching = min(all_matchings, key=score_for_matching)

# Convert into required 2x5 format
result = np.array(best_matching).T

print("Best matching:\n", result)
```
